```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
 sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers = ['o','v'], scatter_kws={'s':40})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, row='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', hue='sex', aspect = 0.6, size = 8)
```
